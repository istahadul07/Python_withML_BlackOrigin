Plots/Graphical representation of my work.
It's done by matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#create some data by numpy
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.title("Sin & Cos wave")

plt.ylabel("Y variable")
plt.plot(x,y,'+') #sin wave
plt.xlabel("X variable")
plt.plot(x,z,'*') #cos wave
plt.show()
```

Create a Parabola

```python
x=np.linspace(-5,10,100)
y=x**2
plt.title("Parabola")
plt.ylabel("Y variable")
plt.plot(x,y)
plt.xlabel("X variable")
plt.show()
```

create a diagram with symbol.

```python
plt.plot(x,y,'*')
plt.show()
```

Bar plot

```python
language=["English","French","Bangla","Latin"]
people=[100,201,300,278]
plt.bar(language,people)
plt.show()
```

Pie Chart

```python
from enum import auto
fig1=plt.figure()
plt.pie(people,labels=language,autopct='%1.1f%%')
plt.show()
```

3D scattered plot

```python
x1=20*np.random.random(100)
y1=np.sin(x1)
z1=np.cos(x1)
fig1=plt.figure()
ax=plt.axes(projection='3d')
ax.scatter(x1,y1,z1)
plt.show()
```
